```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('Resultados.csv')

# AsegÃºrate de que el archivo tiene las columnas esperadas
df
```

```python

# GrÃ¡fico de Barras
plt.figure(figsize=(10, 6))
df['Resultado'].value_counts().plot(kind='bar', color=['green', 'orange'])
plt.title('Frecuencia de Estados de Mensajes')
plt.xlabel('Estado del Mensaje')
plt.ylabel('NÃºmero de Ocurrencias')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# GrÃ¡fico de Pastel
plt.figure(figsize=(8, 8))
df['Resultado'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['green', 'orange', 'lightgreen'])
plt.title('DistribuciÃ³n de Estados de Mensajes')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Contamos los mensajes sin errores y corregidos por cada algoritmo
method_counts = df.groupby(['Metodo', 'Resultado']).size().unstack(fill_value=0)

# Crear grÃ¡fica comparativa de barras
method_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'orange', 'lightblue'])
plt.title('ComparaciÃ³n de DetecciÃ³n vs. CorrecciÃ³n de Errores')
plt.xlabel('Algoritmo')
plt.ylabel('NÃºmero de Mensajes')
plt.xticks(rotation=45)
plt.legend(title='Resultado')
plt.show()
```

```python
# Obtener mÃ©todos Ãºnicos
metodos_unicos = df['Metodo'].unique()

# Lista de colores para cada mÃ©todo
colores = ['lightblue', 'orange', 'lightgreen', 'pink']

# Imprimir mÃ©todos Ãºnicos
print("MÃ©todos Ãºnicos encontrados:")
print(metodos_unicos)

# Crear grÃ¡ficos por cada mÃ©todo
for idx, metodo in enumerate(metodos_unicos):
    # Filtrar el DataFrame por el mÃ©todo actual y limitar a las primeras 100 filas
    df_metodo = df[df['Metodo'] == metodo].head(100)
    
    valores = []
    mensajes = []

    # Recorrer las filas del DataFrame filtrado
    for i, row in df_metodo.iterrows():
        mensajes.append(row['Mensaje'])
        if 'sin errores' in row['Resultado']:
            if not valores:
                valores.append(1)
            else:
                valores.append(valores[-1] + 1)
        else:
            if not valores:
                valores.append(0)
            else:
                valores.append(valores[-1] - 1)

    # Seleccionar un color de la lista, usando el Ã­ndice del mÃ©todo
    color = colores[idx % len(colores)]

    # Plotear los valores para el mÃ©todo actual
    plt.figure(figsize=(12, 6))
    plt.plot(range(1, len(mensajes) + 1), valores, marker='o', color=color) # Color especÃ­fico para cada mÃ©todo
    plt.xlabel('NÃºmero de mensajes')
    plt.ylabel('Ã‰xitos vs Fallos')
    plt.title(f'Mensajes sin errores vs. con errores ({metodo})')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


        
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener mÃ©todos Ãºnicos
metodos_unicos = df['Metodo'].unique()

# Inicializar contadores para Ã©xitos y fallas
exitos = []
fallas = []

# Recopilar datos
for metodo in metodos_unicos:
    df_metodo = df[df['Metodo'] == metodo]
    exitos_count = len(df_metodo[df_metodo['Resultado'].str.contains('sin errores')])
    fallas_count = len(df_metodo[~df_metodo['Resultado'].str.contains('sin errores')])
    exitos.append(exitos_count)
    fallas.append(fallas_count)

# ConfiguraciÃ³n del grÃ¡fico
x = np.arange(len(metodos_unicos))  # posiciones de los mÃ©todos en el eje X
width = 0.35  # ancho de las barras

fig, ax = plt.subplots(figsize=(12, 6))

# Crear barras para Ã©xitos y fallas
rects1 = ax.bar(x - width/2, exitos, width, label='Ã‰xitos', color='lightgreen')
rects2 = ax.bar(x + width/2, fallas, width, label='Fallas', color='salmon')

# AÃ±adir etiquetas y tÃ­tulo
ax.set_xlabel('MÃ©todos')
ax.set_ylabel('NÃºmero de mensajes')
ax.set_title('ComparaciÃ³n de Ã©xitos y fallas por mÃ©todo')
ax.set_xticks(x)
ax.set_xticklabels(metodos_unicos, rotation=45, ha='right')
ax.legend()

# AÃ±adir etiquetas de valores encima de las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```
